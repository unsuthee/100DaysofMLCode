```python
import torch
import torchvision.utils as vutils

def denormalize(img):
    max_val = np.max(img)
    min_val = np.min(img)
    return (img - min_val) / (max_val - min_val)


# Plot a few images sampled from the generator
num_classes = 10

epoch = 50
for digit in range(0, num_classes):
    fn = 'img/generated_img.cvae.{}.epoch{}.npy'.format(digit, epoch)
    
    test_images = np.load(fn)
    vutils.save_image(torch.from_numpy(denormalize(test_images)), 'gen_img.cvae.{}.epoch.{}.png'.format(digit, epoch))    
```
