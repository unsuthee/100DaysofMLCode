```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd

# Plot the loss
log_df = pd.read_csv('loss.cvae.log', sep=',')

ax = log_df['reconstr_error'].plot(legend=True)
log_df['kldiv_error'].plot(ax=ax, legend=True)
ax.set_ylabel('loss')
ax.set_xlabel('parameters update iterations')
```

```python
# Plot a few images sampled from the generator
fig, axes = plt.subplots(figsize=(12,12), nrows=10, ncols=16)

n = 50
for digit in range(0, 10):
    fn = 'img/generated_img.cvae.{}.epoch{}.npy'.format(digit, n)
    print(fn)
    
    test_images = np.load(fn)
    test_images = test_images.squeeze(1)

    for idx in range(16):
        axes[digit, idx].imshow(test_images[idx], cmap='gray')
        axes[digit, idx].axis('off')
```
